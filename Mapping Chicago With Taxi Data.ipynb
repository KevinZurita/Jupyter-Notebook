```python
import pandas as pd
from pandas.io import gbq
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 

df=pd.io.gbq.read_gbq("""SELECT ROUND(pickup_latitude, 4) as lat, ROUND(pickup_longitude, 4) as long, COUNT(*) as num_pickups 
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips` 
WHERE (pickup_latitude BETWEEN 40.00 AND 42.50) AND (pickup_longitude BETWEEN -88.00 AND -87.00 )
GROUP BY lat, long""", project_id ='chicago-project-268502')


```

```python
plt.show(block=True) #Better Styling 

new_style = {'grid': False} #Remove grid
matplotlib.rc('axes', **new_style)
from matplotlib import rcParams
rcParams['figure.figsize'] = (17, 16.5) #Size of figure
rcParams['figure.dpi'] = 250
P=df.plot(kind='scatter', x='long', y='lat',color='cyan',xlim=(-88.00,-87.40),ylim=(41.50, 42.20),s=3,alpha=.6)
P.set_facecolor('black') #Background Color
```
